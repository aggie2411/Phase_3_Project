## Import Data

```python
df = pd.read_csv('../data/train.csv')
validate = pd.read_csv('../data/validate.csv')

df.head()
```

## Split Data

```python
X = df.drop('status_group', axis=1)
y = df['status_group']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Create Pipeline

```python
number_selector = make_column_selector(dtype_include='number', dtype_exclude=['object', 'bool'])
string_selector = make_column_selector(dtype_include=['object', 'bool'])

preprocessing = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), string_selector),
                                        (StandardScaler(), number_selector),
                                       remainder='passthrough')
xgb_pipeline = mp(preprocessing,  SMOTE(random_state=42, n_jobs=-1, sampling_strategy='auto'), \
                   XGBClassifier(random_state=42, n_jobs=-1))
```

## Fit the Data

```python
xgb_pipeline.fit(X_train, y_train)
```

## Make Predictions

```python
y_hat_train = xgb_pipeline.predict(X_train)
y_hat_test = xgb_pipeline.predict(X_test)
```

```python
print(classification_report(y_train, y_hat_train))
```

```python
print(classification_report(y_test, y_hat_test))
```

```python
score = cross_val_score(xgb_pipeline, X_train, y_train, scoring='accuracy', cv=5).mean()
print('5 fold cross-validation score:', score.round(4))
```

## Tune XGBoost

```python
pipe = Pipeline(steps=[('preprocessor', preprocessing),
                         ('sampling', SMOTE(random_state=42, n_jobs=-1, sampling_strategy='minority')),
                        ('classifier', XGBClassifier(random_state=42, n_jobs=-1))
                        ])
```

```python
n_estimators=[50, 100, 500, 1000]
max_depth=[5, 10, 15, 20]
learning_rate=[0.1, 0.3, 0.5]
booster = ['gbtree', 'dart']

parameters = [{'classifier__n_estimators': n_estimators,
              'classifier__max_depth': max_depth,
              'classifier__learning_rate': learning_rate,
              'classifier__booster': booster}]

scorers = {
    'recall_score': make_scorer(recall_score, average='weighted')
     , 'accuracy_score': make_scorer(accuracy_score)
}
```

```python
xgb_randomized = RandomizedSearchCV(pipe, parameters, cv=5, random_state=42, 
                                scoring=scorers, n_jobs=-1, refit='accuracy_score',
                               n_iter=10)
```

```python
xgb_randomized.fit(X_train, y_train)
```

```python
print("Best parameters:", xgb_randomized.best_params_)
print("Best score:", xgb_randomized.best_score_.round(6))
```

## Create New Model with Best Parameters

```python
xgb_pipeline2 = mp(preprocessing,  SMOTE(random_state=42, n_jobs=-1, sampling_strategy='auto'), \
                   XGBClassifier(random_state=42, n_jobs=-1, n_estimators=100, max_depth=15, learning_rate=0.3,\
                                booster='dart'))
```

## Fit the Data

```python
xgb_pipeline2.fit(X_train, y_train)
```

## Generate Predictions

```python
y_hat_train2 = xgb_pipeline2.predict(X_train)
y_hat_test2 = xgb_pipeline2.predict(X_test)
```

```python
print(classification_report(y_train, y_hat_train2))
```

```python
print(classification_report(y_test, y_hat_test2))
```

## Model Validation

```python
score2 = cross_val_score(xgb_pipeline2, X_train, y_train, scoring='accuracy', cv=5).mean()
print('5 fold cross-validation score:', score2.round(5))
```

```python
accuracy2 = accuracy_score(y_test, y_hat_test2)
precision2 = precision_score(y_test, y_hat_test2, average='weighted')
recall2 = recall_score(y_test, y_hat_test2, average='weighted')

# Print the results
print("Accuracy: %.2f%%" % (accuracy2 * 100.0))
print("Precision: %.2f%%" % (precision2 * 100.0))
print("Recall: %.2f%%" % (recall2 * 100.0))
```
